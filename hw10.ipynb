### Euler Method

Here we begin with the Euler Method. This is the least sophisticated method and provides a truly linear, or first order, approximation. A stepcount of 1000 provides oscillatory behavior, but of ever increasing magnitude; which is incorrect. The stepcount of 10000 provides for correct behavior, but with some small yet visible margin of error at the extremas of the function. This is approximately the stepsize at which the behavior is qualitatively correct. The final stepcount of 100000 provides for nearly indistinguishable behavior.
